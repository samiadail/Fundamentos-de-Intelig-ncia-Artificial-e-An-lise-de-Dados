Criar duas listas:
- 5 produtos (apenas inserir valores)
- A primeira lista, custo do produto.
- A segunda lista, o preço de venda.
- Desafio: Calcular o lucro de cada produto.

```python
# NOME DA LISTA
autor_lista = input("Digite o nome do autor da lista: ")
print(autor_lista)


# LISTA DO CUSTO DOS PRODUTOS
print("\nINSIRA O CUSTO DE CADA PRODUTO DOS 5 PRODUTOS")
produto_1_custo = int(input("Digite o custo do primeiro produto: "))
produto_2_custo = int(input("Digite o custo do segundo produto: "))
produto_3_custo = int(input("Digite o custo do terceiro produto: "))
produto_4_custo = int(input("Digite o custo do quarto produto: "))
produto_5_custo = int(input("Digite o custo do quinto produto: "))
lista_custo = [produto_1_custo, produto_2_custo, produto_3_custo, produto_4_custo, produto_5_custo]
print("Lista dos custos dos 5 produtos: ")
print(lista_custo)

#LISTA DO PREÇO DE VENDA DOS PRODUTOS
print("\nINSIRA O VALOR DE VENDA DE CADA PRODUTO DOS 5 PRODUTOS")
produto_1_venda = int(input("Digite o valor de venda primeiro produto: "))
produto_2_venda = int(input("Digite o valor de venda segundo produto: "))
produto_3_venda = int(input("Digite o valor de venda terceiro produto: "))
produto_4_venda = int(input("Digite o valor de venda quarto produto: "))
produto_5_venda = int(input("Digite o valor de venda quinto produto: "))
lista_preco_venda = [produto_1_venda, produto_2_venda, produto_3_venda, produto_4_venda, produto_5_venda]
print("Lista do valor de venda dos 5 produtos: ")
print(lista_preco_venda)

lista_custo = np.array(lista_custo)
lista_preco_venda = np.array(lista_preco_venda)

faturamento = lista_preco_venda - lista_custo
print(f"\nO Lucro de cada produto é: {faturamento}") 
```

```python
# Em milhar
salarios = np.array([2, 3, 4, 5, 6, 500])
maior_salario = np.max(salarios)
print(maior_salario)
```

```python
media_salarios = np.mean(salarios)
print(f"{media_salarios:.2f}")
```

```python
mediana_salarios = np.median(salarios)
print(mediana_salarios)
```

```python
desvio = np.std(salarios)
print(desvio)
```

Criar uma lista de vendas por dia
vendas = [10, 20, 45, 5, 90]
- Exibir a maior venda
- Exibir a menor venda
- Exibir a média de vendas
- Exibir a mediana de vendas

```python
vendas = [10, 20, 45, 5, 90]

maior_venda = np.max(vendas)
print(maior_venda)

menor_vendas = np.min(vendas)
print(menor_vendas)

media_vendas = np.mean(vendas)
print(media_vendas)

mediana_vendas = np.median(vendas)
print(mediana_vendas)
```
