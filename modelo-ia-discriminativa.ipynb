```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r"data\dataset_aula03_student_habits_performance_2.csv")
sns.lmplot(data=df, x="study_hours_per_day", y="exam_score");
```

Separando os dados de entrada com os dados que queremos descobrir.

```python
X = df[["study_hours_per_day",
        "sleep_hours",
        "netflix_hours",
        "exercise_frequency",
        "age",
        "attendance_percentage"
        ]]

y = df["exam_score"]
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(X, y)
```

Definindo o tanto de horas/porcentagem

```python
novo_aluno = pd.DataFrame({
    "study_hours_per_day":[6],
    "sleep_hours":[7],
    "netflix_hours":[1],
    "exercise_frequency":[2],
    "age":[40],
    "attendance_percentage":[100]
    })
```

```python
resultado = modelo.predict(novo_aluno)
print(resultado)
```

```python
# Coeficient Angular
modelo.coef_
```

```python
# Coeficiente Linear
modelo.intercept_
```

```python
modelo.score(X, y)
```
