Verificar a qualidade das notas de um aluno
Maior do que 9: Excelente
Maior do que 7: Boa
Maior do que 5: Regular
Abaixo de 5: Reprovado

```python
# O comando input solicita a interação do usuário
nota = int(input("Digite a sua nota"))
print(nota)
```

```python
if nota > 9:
    print("NOTA EXCELENTE")
elif nota > 7:
    print("NOTA BOA")
elif nota > 5:
    print("NOTA REGULAR")
else:
    print("REPROVADO")
```
