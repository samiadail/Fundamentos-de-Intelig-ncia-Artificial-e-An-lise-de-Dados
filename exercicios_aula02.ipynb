```python
# Exercicio 1
# Importe as bibliotecas pandas e seaborn e configure o tema visual dos graficos.
import pandas as pd
import seaborn as sns

```

```python
# Exercicio 2
# Carregue o arquivo modafina_vendas.csv em um DataFrame
# e mostre as primeiras linhas para conhecer os dados.
df = pd.read_csv(r"data\dataset_aula02_exercicio_modafina (1).csv", parse_dates=["data_venda"])
df.head()
```

```python
# Exercicio 3
# Mostre quantas linhas e quantas colunas a base de dados possui.
df.shape
```

```python
# Exercicio 4
# Mostre as informacoes gerais da base: o nome das colunas,
# o tipo de cada uma e a quantidade de valores nao nulos.
df.info()
df.notnull().sum()
```

```python
# Exercicio 5
# Crie uma nova coluna chamada faturamento, multiplicando o preco
# pela quantidade de cada venda. Depois, mostre as primeiras linhas.
df["faturamento"] = df["preco"] * df["quantidade"]
df.head()
```

```python
# Exercicio 6
# Calcule o faturamento total da loja (a soma da coluna faturamento).
# Em seguida, exiba esse valor formatado como moeda brasileira,
# no formato R$ 0.000,00 (exemplo: R$ 232.687,92).

faturamento_total = df["faturamento"].sum()
print(faturamento_total)

faturamento_formatado = f"{faturamento_total:,.2f}".replace(",","x").replace(".",",").replace("x",".")
print(f"O faturamento total é R$ {faturamento_formatado}")
```

```python
# Exercicio 7
# Descubra quantos pedidos existem de cada categoria de produto.
# Conta a frequência de cada categoria na coluna 'categoria_produto'
contagem_categorias = df["categoria_produto"].value_counts()

# Exibe o resultado
print(contagem_categorias)
```

```python
# Exercicio 8
# Descubra quanto cada categoria de produto faturou no total,
# ordenando da que mais faturou para a que menos faturou.

# Agrupa por categoria e soma o faturamento, depois ordena de forma decrescente
faturamento_por_categoria = df.groupby("categoria_produto")["faturamento"].sum().sort_values(ascending=True)

# Exibe o resultado
print(faturamento_por_categoria)
```

```python
def converter(valor):
    return f"R$ {valor:,.2f}".replace(",", "x").replace(".", ",").replace("x", ".")

faturamento_por_categoria = df.groupby("categoria_produto")["faturamento"].sum().sort_values(ascending=True)


faturamento_por_categoria.map(converter)
```

```python
# Exercicio 9
# Crie um grafico que mostre a quantidade de pedidos por categoria de produto.
# Coloque um titulo no grafico e nomes nos eixos.

# -----------REPOSTA DO ANONIMO-----------

import matplotlib.pyplot as plt

# Cria o gráfico de barras com a contagem de pedidos por categoria
plt.figure(figsize=(10, 6)) # Define o tamanho da figura
sns.countplot(data=df, x="categoria_produto", palette="viridis")

# Adiciona título e nomes aos eixos
plt.title("Quantidade de Pedidos por Categoria de Produto", fontsize=15)
plt.xlabel("Categoria do Produto", fontsize=12)
plt.ylabel("Quantidade de Pedidos", fontsize=12)

# Exibe o gráfico
plt.show()

# -----------REPOSTA DO ANONIMO-----------


# 06/07/2026 RESPOSTA DO PROFESSOR
grafico = sns.countplot(data=df, x=df["categoria_produto"], order=df["categoria_produto"].value_counts().index);
grafico.bar_label(grafico.containers[0]);
grafico.set_title("Vendas por categoria");
grafico.set_xlabel("Categorias");
grafico.set_ylabel("Quantidade");
# 06/07/2026 RESPOSTA DO PROFESSOR
```

```python
# Exercicio 10
# Crie um grafico que compare a distribuicao do preco entre as
# diferentes categorias de produto. Coloque um titulo no grafico.

# -----------REPOSTA DO ANONIMO-----------
import matplotlib.pyplot as plt

# Define o tamanho da figura
plt.figure(figsize=(12, 6))

# Cria o boxplot
sns.boxplot(data=df, x="categoria_produto", y="preco")

# Adiciona o título
plt.title("Distribuição de Preço por Categoria de Produto", fontsize=15)
plt.xlabel("Categoria do Produto")
plt.ylabel("Preço (R$)")

# Exibe o gráfico
plt.show()
# -----------REPOSTA DO ANONIMO-----------


# 06/07/2026 RESPOSTA DO PROFESSOR
grafico = sns.boxplot(data=df, x="categoria_produto", y="preco");
grafico.set_title("Dstribuição de preços por categoria");
grafico.set_xlabel("Categorias");
grafico.set_ylabel("Preços");
# 06/07/2026 RESPOSTA DO PROFESSOR
```
