# Atividade: Construindo um relatório com IA

Nesta segunda parte da aula, você vai usar uma inteligência artificial (Claude, ChatGPT ou Gemini) para aprofundar a análise que fizemos no notebook. Ao final, você terá um relatório completo, montado neste próprio notebook, em células de texto.

**Antes de começar:**

1. Abra uma conta gratuita em claude.ai, chat.openai.com ou gemini.google.com
2. Envie o arquivo `student_habits_performance.csv` no chat da IA
3. Neste notebook, crie uma célula de texto com o título: **Relatório de Análise — Hábitos e Desempenho dos Alunos**

> **Observação importante:** esta base é fictícia, pode enviar sem problemas. No trabalho real, nunca envie dados de pessoas reais para uma IA pública.

---

## Parte 1

Faça as três perguntas abaixo, na ordem, para a mesma IA. Observe como a resposta melhora a cada versão.

**Pergunta 1 (apenas a pergunta):**

```
Qual hábito mais influencia a nota dos alunos?
```

**Pergunta 2 (a pergunta com contexto):**

```
Esta base tem 1000 alunos, com hábitos (horas de estudo, sono, redes sociais, exercício) e a nota final no exame (exam_score). Analise quais hábitos mais influenciam a nota e mostre os números que comprovam.
```

**Pergunta 3 (contexto, persona e formato de resposta):**

```
Você é um analista de dados educacionais experiente. Esta base tem 1000 alunos, com hábitos e a nota final (exam_score). Identifique os 3 hábitos que mais impactam a nota, apresente cada um com o número que comprova e escreva em linguagem clara para a direção de uma escola.
```

No relatório, crie a seção **"1. Engenharia de Prompt"** e responda: o que mudou da primeira para a terceira resposta? Qual delas você usaria num trabalho real? Cole a melhor das três respostas.

---

## Parte 2

O heatmap que fizemos usou apenas as colunas numéricas. A base também tem colunas de texto que o código não analisou. Use o estilo da Pergunta 3 (com contexto e persona) para investigar pelo menos três das perguntas abaixo:

```
A qualidade da dieta do aluno (diet_quality) faz diferença na nota? Mostre a média por grupo.
```

```
Alunos com trabalho de meio período (part_time_job) vão pior nas notas?
```

```
A escolaridade dos pais (parental_education_level) tem relação com o desempenho?
```

```
Existe uma quantidade ideal de horas de sono para tirar as melhores notas?
```

No relatório, crie a seção **"2. Novas análises"** e cole os insights que achou mais interessantes.

**Atenção:** alguns desses fatores quase não mudam a nota. Isso também é uma descoberta, pois mostra que nem tudo que imaginamos ser importante realmente é.

---

## Parte 3

Faça a seguinte pergunta para a IA:

```
Por que os alunos mais velhos desta base tiraram notas muito mais altas?
```

Agora confira no notebook: a idade tem relação com a nota? Observe como a IA respondeu. Ela inventou uma explicação ou avisou que não havia relação nos dados?

No relatório, crie a seção **"3. Cuidados e Ética no uso da IA"** e escreva, com suas palavras, o que aconteceu e por que é perigoso confiar cegamente na IA no trabalho.

Antes de finalizar, confira um dos números que a IA forneceu em qualquer uma das partes, usando o seu notebook. Um bom analista sempre confere.
